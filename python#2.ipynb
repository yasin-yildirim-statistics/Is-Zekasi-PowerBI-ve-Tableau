```python
# len() : Bir dizinin (string, liste, demet, sözlük vb.) uzunluğunu döndüren yerleşik bir Python fonksiyonudur.
metin = "Merhaba Dünya"
uzunluk = len(metin)
print("Metin uzunluğu:", uzunluk)
```

```python
my_list = [1,2,3,'A','B','C',True,False]
liste_uzunlugu = len(my_list)
print("Liste uzunluğu:", liste_uzunlugu)
```

```python
# sum() : Bir dizideki sayısal değerlerin toplamını döndüren fonksiyonudur.
liste1 = [10,2,3,4,1,5,6,21,14,14,25,36]
print("Listenin toplamı:", sum(liste1))
```

```python
min_deger = min(liste1)
print("Listenin en küçük değeri:", min_deger)
```

```python
print(max(liste1))
```

```python
# aritmetik ortalamasını hesaplayın
toplam = sum(liste1)
eleman_sayisi = len(liste1)
ortalama = toplam / eleman_sayisi
print("Listenin aritmetik ortalaması:", ortalama)
```

```python
# liste fonksiyonları:
# append(veri) : liste sonuna eleman ekler.
```

```python
teams = ['Fenerbahçe', 'Galatasaray', 'Beşiktaş', 'Trabzonspor']
```

```python
print(teams)
```

```python
teams.append('Başakşehir')
print(teams)
```

```python
# kullanıcıdan kaç adet takım girilmesi isteniyorsa o bilgiyi alalım ve o kadar takım ismi alıp listeye ekleyelim.
takım_sayisi = int(input("Kaç adet takım eklemek istiyorsunuz? "))
for i in range(takım_sayisi):
    takım_adi = input('Takım Adı : ')
    teams.append(takım_adi)
```

```python
print(teams)
```

```python
# remove(veri) : listeden eleman siler.
teams.remove('Başakşehir')
print(teams)
```

```python
# sort artan veya a'dan z'ye sıralama işlemleri yapar.
teams.sort()
```

```python
print(teams)
```

```python
teams.sort(reverse=True)
print(teams)
```
