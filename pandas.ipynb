```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Öğrenci': ['Ayşe', 'Fatma', 'Ali', 'Ömer', 'Buse', 'Evren', 'Zeynep'],
    'Matematik': [85, 65, 35, 92, 88, 76, 95],
    'Fizik': [42, 51, 38, 64, 75, 30, 15],
    'Yaş': [17, 18, 17, 18, 17, 19, 19],
    'Sınıf': ['11A','11B','11C','11B','11B','11A','11C'] }
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
# head() fonksiyonu ilk 5 veriyi listeler
df.head(3)
```

```python
# tail() fonksiyonu sondan ilk 5 veriyi listeler
df.tail(2)
```

```python
df.describe()
```

```python
# sütun seçme
df['Matematik']
```

```python
# birden fazla kolon seçme işlemi
df[['Öğrenci','Matematik','Fizik']]
```

```python
# matematik notu 70'ten büyük olan öğrencileri listeleme
df[df['Matematik'] > 70]
```

```python
# birden fazla kolon için head kullanma
df[['Öğrenci','Matematik','Yaş']].head(4)
```

```python
# satır bazında filtreleme
# matematik puanları 80'den falza olan öğrenci bilgierlini listeleyiniz.
# select * from öğrenciler where matematik > 80

df[df['Matematik'] > 80]
```

```python
# matematik dersinden 70'den fazla alan öğrenciler ve aynı zamanda fizik dersinden 50'den fazla alan öğrencileri listeleyiniz.
# and -> & , or -> |
df[(df['Matematik'] > 70) & (df['Fizik'] > 50)]
```

```python
#11A sınıfındaki öğrencileri listeleyiniz.
df[df['Sınıf'] == '11A']
```

```python
# kolon ekleme
df['Ortalama'] = (df['Matematik'] + df['Fizik']) / 2
df
```

```python
df['Cinsiyet'] = ['K','K','E','E','K','E','K']
df
```

```python
# ortalama sütununa göre artan bir şekilde sıralama
df.sort_values('Ortalama')
```

```python
# azalan bir şekilde sıralama
df.sort_values('Ortalama', ascending=False)
```

```python
df.sort_values('Öğrenci', ascending=False)
# asc -> artan
# desc -> azalan
```

```python
# gruplama
# mean() count() min() max() sum() gibi fonksiyonlar kullanılabilir.
```

```python
# sınıflara göre ortalama hesaplama
df.groupby('Sınıf')['Ortalama'].mean()
```

```python
# sınıflara göre öğrenci sayısı
df.groupby('Sınıf')['Öğrenci'].count()
```

```python
# cinsiyete göre öğrenci sayısı
df.groupby('Cinsiyet')['Öğrenci'].count()
```

```python
df.groupby('Yaş')['Ortalama'].mean()
```

```python
# yaş alanına göre not ortalaması en büyük olan notları listeleyiniz.
df.groupby('Yaş')['Ortalama'].max()
```

```python
# en başarılı öğrenci
df.loc[df['Ortalama'].idxmax()]
```

```python
# matematikte en başarılı öğrenci
df.loc[df['Matematik'].idxmax()]
```

```python
df['Ortalama'].idxmax()
```

```python
df['Ortalama'].idxmin()
```

```python
# en başarısız öğrenci
df.loc[df['Ortalama'].idxmin()]
```

```python
# matematikte en başarısız öğrenci
df.loc[df['Matematik'].idxmin()]
```

```python
# sınıf bazında başarı sıralaması
df.groupby('Sınıf')['Ortalama'].mean().sort_values(ascending=False)
```

```python
# excell' akatrma
exc1_data = df.sort_values('Öğrenci', ascending=False)
exc1_data
```

```python
exc1_data.to_excel('ogrenci_notları.xlsx')
```
