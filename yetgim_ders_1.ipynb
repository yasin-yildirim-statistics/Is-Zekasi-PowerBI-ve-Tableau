```python
# mantıksal operatörler :
# and , or , not
durum1 = (3 <5 and 5>8)
# False =   True ve  False

durum2 = (25 == 35 and 85 != 85)
#False =   False   ve   False

durum3 = 'Ömer' == 'Sadullah' and 'B' > 'C'
#False      =   False              ve   false

durum4 = (3 <5 or 5>8)
#True =   True veya False
durum5 = (25 == 35 or 85 != 85)
#False=    False  veya False

durum6 = 'Ömer' == 'Sadullah' or not 'B' > 'C'
#True=         False    veya True

print(durum1)
print(durum2)
print(durum3)
print(durum4)
print(durum5)
print(durum6)
```

```python
# ternary operatör
age = int(input('Lütfen Yaşınızı giriniz.'))
result = 'Ehliyet alabilirsiniz.' if age >= 18 else 'Ehliyet alamazsınız.'
print(result)
```

```python
# Kullanıcıdan sıcaklık bilgisini alınız.
# sıcaklık 0 dan küçük gelmesi durumunda Hava Soğuk ekran çıktısı versin
# 0 ile 20 arasında Ilık
# 20 ile 40 arasında sıcak
# 40 dan fazla gelir ise Çok sıcak
```

```python
#  0 <=  sıcaklık <20

sicaklik = int(input('Lütfen Sıcaklık giriniz C(celcius) cinsinden : '))
if sicaklik <0:
  print('Hava : Soğuk')
elif sicaklik>=0 and sicaklik <20:
  print('Hava : Ilık')
elif sicaklik>=20 and sicaklik<40:
  print('Hava: Sıcak')
else:
  print('Hava : Çok Sıcak')
```

```python
sicaklik = int(input('Lütfen Sıcaklık giriniz C(celcius) cinsinden : '))
if sicaklik <0:
  print('Hava : Soğuk')
elif 0<sicaklik<20:
  print('Hava : Ilık')
elif 20<=sicaklik<40:
  print('Hava: Sıcak')
elif sicaklik >=40:
  print('Hava : Çok Sıcak')
```

```python
# liste içerisinde ki sayıları teker teker kontrol eden ve ilgili sayının tek mi çift mi olduğunu bulan programı yazınız.
liste = [15,21,78,54,23,44,34,58,41,39,55,210,147,258,369]
for number in liste:
  if number % 2 ==0:
    print('Çift sayı : ',number)
  else:
    print('Tek Sayı : ',number)
```

```python
# liste içerisinde 100 den büyük bir sayı gelir ise döngüyü durdur.
for number in liste:
  if number > 100:
    break
  print(number)
```

```python
# liste içerisinde Çift sayıları gösteriniz.
# Not : Tek sayıları atlasın

for number in liste:
  if number % 2 ==1:
    continue
  print(number)
```
