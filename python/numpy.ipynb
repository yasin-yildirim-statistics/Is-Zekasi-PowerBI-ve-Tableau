```python
#saf python ile kodlama
baslangic = time.time()
A = list(range(100000001))
B = list(range(100000001))
C = []
for i in range(len(B)):
    toplam = A[i] + B[i]
    C.append(toplam)
bitis = time.time()
sure = bitis - baslangic
print("Saf Python ile çalışma süresi: ", sure)
```
